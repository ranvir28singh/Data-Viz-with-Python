## More Matplotlib (and Seaborn) !

Now that we know Matplotlib, let's use our skills to visualize data from our previous dataset on unemployment. Let's follow the trend in unemployment rates for four counties: Autauga, AL, Hartford, CT, Carlisle, KY and Dallas, TX. We want to find any trend in rates of unemployment across the years (2007 to 2017). 

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
unemployment_df = pd.read_excel('https://www.ers.usda.gov/webdocs/DataFiles/48747/Unemployment.xls?v=0', header = 7)
unemployment_df.head()
unemployment_df = unemployment_df.dropna()
```

```python
df_2 = unemployment_df[['Area_name','Unemployment_rate_2007','Unemployment_rate_2008','Unemployment_rate_2009',
       'Unemployment_rate_2010','Unemployment_rate_2011','Unemployment_rate_2012','Unemployment_rate_2013',
       'Unemployment_rate_2014','Unemployment_rate_2015','Unemployment_rate_2016','Unemployment_rate_2017']]
```

```python
fig, ax = plt.subplots(2,2, figsize = (16,16), sharex = True, sharey = True);

ax[0,0].plot(df_2[df_2['Area_name'] == 'Autauga County, AL'].iloc[:,1:].columns,
              df_2[df_2['Area_name'] == 'Autauga County, AL'].iloc[:,1:].values[0],
                label = 'Autauga, AL');
ax[0,0].set_ylabel('Unemployment %')

ax[0,1].plot(df_2[df_2['Area_name'] == 'Hartford County, CT'].iloc[:,1:].columns,
              df_2[df_2['Area_name'] == 'Hartford County, CT'].iloc[:,1:].values[0],
                label = 'Hartford, CT');
ax[1,0].plot(df_2[df_2['Area_name'] == 'Carlisle County, KY'].iloc[:,1:].columns,
              df_2[df_2['Area_name'] == 'Carlisle County, KY'].iloc[:,1:].values[0],
                label = 'Carlisle, KY');
ax[1,0].set_ylabel('Unemployment %')
ax[1,1].plot(df_2[df_2['Area_name'] == 'Dallas County, TX'].iloc[:,1:].columns,
              df_2[df_2['Area_name'] == 'Dallas County, TX'].iloc[:,1:].values[0],
                label = 'Dallas, TX');

columns = [x[18:] for x in df_2[df_2['Area_name'] == 'Autauga County, AL'].iloc[:,1:].columns]
ax[0,0].set_xticklabels(labels = columns, rotation = 'vertical');
ax[0,0].legend();
ax[0,1].set_xticklabels(labels = columns, rotation = 'vertical');
ax[0,1].legend();
ax[1,0].set_xticklabels(labels = columns, rotation = 'vertical');
ax[1,0].legend();
ax[1,1].set_xticklabels(labels = columns, rotation = 'vertical');
ax[1,1].legend();
```

We can also visualize the data in one plot as follows:

```python
plt.figure(figsize = (16,6))
plt.plot(columns,
              df_2[df_2['Area_name'] == 'Autauga County, AL'].iloc[:,1:].values[0],
                label = 'Autauga, AL');
plt.plot(columns,
              df_2[df_2['Area_name'] == 'Hartford County, CT'].iloc[:,1:].values[0],
                label = 'Hartford, CT');
plt.xticks([x for x in columns], [x for x in range(2007, 2018)])
plt.plot(columns,
              df_2[df_2['Area_name'] == 'Carlisle County, KY'].iloc[:,1:].values[0],
                label = 'Carlisle, KY');
plt.xticks([x for x in columns], [x for x in range(2007, 2018)])
plt.plot(columns,
              df_2[df_2['Area_name'] == 'Dallas County, TX'].iloc[:,1:].values[0],
                label = 'Dallas, TX');
plt.ylabel('Unemployment %')

plt.legend();
```

## Seaborn

Look at the distribution for rates of unemployment across all counties for all years. Do this using Violinplots and Boxplots. What can you tell about the distributions across the years from 2007 to 2017?

#### Violinplot

```python
import seaborn as sns

plt.figure(figsize = (16,8));
pal = sns.cubehelix_palette(11, rot=-.5, dark=.3);
unemp_dist = sns.violinplot(data = df_2.iloc[:,1:], palette=pal, inner="points");
unemp_dist.set_xticklabels(labels = df_2.iloc[:,1:].columns, rotation=90);
plt.title("Unemployment Rate for All Counties by Year")
```

#### Boxplot

```python
plt.figure(figsize = (16,8));
pal = sns.cubehelix_palette(11, rot=-.5, dark=.3);
unemp_dist = sns.boxplot(data = df_2.iloc[:,1:], palette=pal);
unemp_dist.set_xticklabels(labels = df_2.iloc[:,1:].columns, rotation=90);
plt.title("Unemployment Rate for All Counties by Year");
```

A variation of the boxplot

```python
plt.figure(figsize = (16,8));
pal = sns.cubehelix_palette(11, rot=-.5, dark=.3);
unemp_dist = sns.boxenplot(data = df_2.iloc[:,1:], palette=pal);
unemp_dist.set_xticklabels(labels = df_2.iloc[:,1:].columns, rotation=90);
plt.title("Unemployment Rate for All Counties by Year");
```
