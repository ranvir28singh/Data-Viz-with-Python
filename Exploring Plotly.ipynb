```python
import pandas as pd
import plotly
import matplotlib.pyplot as plt
import seaborn as sns
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from IPython.core.display import display, HTML
%matplotlib inline
import numpy as np
import plotly.graph_objs as go
```

```python
plotly.offline.init_notebook_mode(connected=True)
```

```python
sales_df=pd.read_excel('sales.xls','Orders')
```

```python
sales_df.head()
```

```python
trace1=go.Scatter(x=sales_df['Sales'],y=sales_df['Profit'],name='Profit')
#trace2=go.Scatter(x=sales_df['Sales'],y=sales_df['Discount'],mode='markers',name='Discount')
data=[trace1]
layout=go.Layout(title='Sales vs Profit', xaxis=dict(title='Sales'),yaxis=dict(title='Profit'),hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
trace1=go.Scatter(x=sales_df['Sales'],y=sales_df['Profit'],mode='markers',name='Profit')
trace2=go.Scatter(x=sales_df['Sales'],y=sales_df['Discount'],mode='markers',name='Discount')
data=[trace1,trace2]
layout=go.Layout(title='Sales vs Profit & Discount', xaxis=dict(title='Sales'),yaxis=dict(title='Profit'),hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
order_pri=pd.pivot_table(sales_df,values=['Sales','Profit'],index='Order Priority',aggfunc='mean')
```

```python
order_pri
```

```python
trace1=go.Bar(x=order_pri.index,y=order_pri['Sales'],name='Sales')
#trace2=go.Bar(x=order_pri.index,y=order_pri['Profit'],name='Profit')
data=[trace1]
layout=go.Layout(title='Avg Sales by Order Priority',hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
trace1=go.Bar(x=order_pri.index,y=order_pri['Sales'],name='Sales',text=order_pri['Sales'],textposition='auto')
trace2=go.Bar(x=order_pri.index,y=order_pri['Profit'],name='Profit',text=order_pri['Profit'])
data=[trace1,trace2]
layout=go.Layout(title='Avg Sales & Profit by Order Priority',hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
trace1=go.Bar(x=order_pri.index,y=order_pri['Sales'],name='Sales')
trace2=go.Bar(x=order_pri.index,y=order_pri['Profit'],name='Profit')
data=[trace1,trace2]
layout=go.Layout(title='Avg Sales & Profit by Order Priority',hovermode='closest',barmode='stack')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
trace1=go.Scatter(x=sales_df['Sales'],y=sales_df['Profit'],mode='markers',marker=dict(size=10,color='darkorchid',symbol='square'))
#trace2=go.Scatter(x=sales_df['Sales'],y=sales_df['Discount'],mode='markers',name='Discount')
data=[trace1]
layout=go.Layout(title='Sales vs Profit & Discount', xaxis=dict(title='Sales'),yaxis=dict(title='Profit'),hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
years_df=pd.pivot_table(country_df,values='pop',index='year',columns='continent',aggfunc='sum')
years_df.head()
```

```python
years_df.index
```

```python
years_df.loc[:,'Africa']
```

```python
trace1=go.Scatter(y=years_df.loc[:,'Americas'],x=years_df.index,name='Americas',mode='lines')
trace2=go.Scatter(y=years_df.loc[:,'Asia'],x=years_df.index,name='Asia',mode='lines')
data=[trace1,trace2]
layout=go.Layout(title='Population by Years', xaxis=dict(title='Population'),yaxis=dict(title='Year'),hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```

```python
plot(figure, filename='popbyyear.html')
```

```python
trace1=go.Scatter(y=years_df.loc[:,'Americas'],x=years_df.index,name='Americas',mode='markers+lines+text')
trace2=go.Scatter(y=years_df.loc[:,'Asia'],x=years_df.index,name='Asia',mode='markers+lines+text')
data=[trace1,trace2]
layout=go.Layout(title='Population by Years', xaxis=dict(title='Population'),yaxis=dict(title='Year'),hovermode='closest')
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```
